```python
import plotly.graph_objects as go
import numpy as np

def q_triangle(x1, x2):
    return 1/2 - abs(x2 - 1/2)

def compute_minimal_graph():
    vertices = [[0, 0, 0], [0.5, 0, 0.5], [1, 0, 0], [0, 1, 0], [0.5, 1, 0.5], [1, 1, 0]]
    triangles = [[0, 0.5, 4], [1, 2, 4], [2, 3, 4], [0, 4, 3], [0, 3, 5]]

    def minimal_graph(x1, x2):
        q_values = []
        for i, triangle in enumerate(triangles):
            in_triangle = (x1 >= vertices[triangle[0]][0]) & (x1 <= vertices[triangle[1]][0]) & \
                          (x2 >= vertices[triangle[0]][1]) & (x2 <= vertices[triangle[2]][1])
            q_values.append(np.where(in_triangle, q_triangle(x1, x2), 0))
        return sum(q_values)

    return minimal_graph

# Example usage
minimal_graph_fn = compute_minimal_graph()

# Create grid points for surface plotting
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate the minimal graph function at grid points
Z = minimal_graph_fn(X1, X2)

# Create the surface plot using Plotly
fig = go.Figure(data=go.Surface(x=X1, y=X2, z=Z))

# Set the layout options
fig.update_layout(
    title='Minimal Triangular Graph',
    scene=dict(
        xaxis_title='x1',
        yaxis_title='x2',
        zaxis_title='q(x1, x2)',
    )
)

# Show the figure
fig.show()
```
