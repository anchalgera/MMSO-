Explicit one step euler method 
Differential Equation y'=x
Formulae : y(t+h)=t(t)+h*f(x,y)

finding numerical solution 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# defining the function y'=f(x,y) and f(x,y)=x
def fun(x,y):
    return x
```

```python
# defining initial conditions i.e. x0=0,y0=0 and step size =0.05 
x,y,h=0.0,100,0.1
```

```python
a=[]
b=[]
```

```python
for i in range(10):
    y = y +  h*fun(x, y)
    x = x + h
    a.append(x)
    b.append(y)
```

finding exact solution

```python
x,y=0.0,100
```

```python
# defining the function y'=f(x,y) and f(x,y)=x
def y_exact(x,y):
    return (-20)*(y)
```

```python
c=[]
d=[]
```

```python
for i in range(1000):
    y = y_exact(x,y)
    x = x + h
    c.append(x)
    d.append(y)
```

```python
error=np.subtract(b,d)
```

```python
fig,ax=plt.subplots()
ax.plot(a,b,'-r',label='explicit one step euler method ')
ax.plot(c,d,'--b',label='exact solution')
leg = ax.legend();
```

```python
fig,ax=plt.subplots()
ax.plot(error,label="error")
leg = ax.legend()
```
