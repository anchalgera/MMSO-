Runga kutta of 4th order

finding numerical solution
Differential Equation y'=x
formulae  k1 = h*fun(x, y)
          k2 = h*fun(x+h/2, y+k1/2)
          k3 = h*fun(x+h/2, y+k2/2)
          k4 = h*fun(x+h, y+k3)
          y = y + 1/(6.0)*(k1+2*k2+2*k3+k4)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
# defining the function y'=f(x,y) and f(x,y)=x
def fun(x,y):
    return x
```

```python
# defining initial conditions i.e. x0=0,y0=0 and step size =0.05 
x,y,h=0.0,0.0,0.05
```

```python
a=[]
b=[]
```

```python
for i in range(1000):
    k1 = h*fun(x, y)
    k2 = h*fun(x+h/2, y+k1/2)
    k3 = h*fun(x+h/2, y+k2/2)
    k4 = h*fun(x+h, y+k3)
    y = y + 1/(6.0)*(k1+2*k2+2*k3+k4)
    x = x + h
    a.append(x)
    b.append(y)
```

finding exact solution

```python
x,y=0.0,0.0
```

```python
# defining the function y'=f(x,y) and f(x,y)=x
def y_exact(x,y):
    return (0.5)*(x*x)
```

```python
c=[]
d=[]
```

```python
for i in range(1000):
    y = y_exact(x,y)
    x = x + h
    c.append(x)
    d.append(y)
```

```python
error=np.subtract(b,d)
```

```python
fig,ax=plt.subplots()
ax.plot(a,b,'-r',label='runga kutta 4th order')
ax.plot(c,d,'--b',label='exact solution')
leg = ax.legend();
```

```python
fig,ax=plt.subplots()
ax.plot(error,label="error")
leg = ax.legend()
```
